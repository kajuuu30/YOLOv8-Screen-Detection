```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="/content/Kajal_Screen/data.yaml",  # ğŸ‘ˆ ye sahi path hai
    epochs=100,
    imgsz=840,
    batch=16
)
```

```python
from ultralytics import YOLO
import shutil

# 1. Apna trained model Drive se load karo
model = YOLO("/content/drive/MyDrive/YOLO_img/best.pt")

# 2. Test images folder par prediction
results = model.predict(
    source="/content/images_mobile_phone",
    save=True,
    conf=0.5
)

# 3. Output ko Drive me save kar do
output_path = str(results[0].save_dir)  # prediction folder
drive_path = "/content/drive/MyDrive/YOLO_img/YOLO_predictions"

shutil.copytree(output_path, drive_path, dirs_exist_ok=True)

print("âœ… Predictions saved permanently to:", drive_path)
```
